```python
from transformers import pipeline
from youtube_transcript_api import YouTubeTranscriptApi
```

```python
# using Bert model: encoder-encoder (seq2seq) model
summarizer = pipeline("summarization", model="facebook/bart-large-cnn")
```

```python
youtube_video = "https://www.youtube.com/watch?v=Tuw8hxrFBH8"  # Steve Jobs motivation speach
id = youtube_video.split("=")[1]
```

```python
# extracting text from each "meta-data- part"
transcript_lines = YouTubeTranscriptApi.get_transcript(id)

transcript = ""
for i in transcript_lines:
  transcript +="" + i["text"]

print(len(transcript))
```

```python
# the transcript is long for the summarization with Bert. Process in chunks
chunks= int(len(transcript)/1000)

summary= []
for i in range(0,chunks+1):

  start = 0
  start= i* 1000
  end= (start +1000)
  summary_text = summarizer(transcript[start:end], max_length=100) # (summarizer(transcript ,max_length=300, min_length=20, do_sample=False))
  summary_text = summary_text[0]
  summary_text = summary_text['summary_text']
  summary.append(summary_text)


print("Summary of the video:" + summary)
# len(str(summary))
```
